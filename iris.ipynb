```python
#Statistics of this dataset
stats = iris.describe()
stats
```

```python
#calculate range, and move columns 25%,50%,75%  and mean values into a dataframe
stats.loc['range'] = stats.loc['max'] - stats.loc['min']
new_df = stats.loc[['mean', 'range', '25%', '50%', '75%']]
new_df.rename({'50%': 'median'}, inplace=True)
new_df
```

```python
#Calculate mean and median or others using multiple functions at once - 2 methods

iris.groupby('Species').agg(['mean', 'median'])  # passing a list of recognized strings
iris.groupby('Species').agg([np.mean, np.median])
```

```python
#If we want an explicit aggr function like max for a column
agg_dict={field: ['mean', 'median'] for  field in iris.columns.tolist() if field != 'Species'}


agg_dict['Petal_Length'] = 'max'
iris.groupby('Species').agg(agg_dict)
```

```python
# Count the number of occurrences of each species
species_counts = iris['Species'].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))
my_ex = [0.4, 0, 0.1]
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%',explode=my_ex, startangle=140)
plt.title('Distribution of Iris Species')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the pie chart
plt.show()
```
